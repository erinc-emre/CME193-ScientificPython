## 9 Numpy

```python
import numpy as np
import scipy as sp

n = 200
m = 500

k = np.random.rand(m)
l = np.random.rand(m)

A = np.random.normal(0,1, (n,m))
B = sp.linalg.toeplitz(k,l)

print('Matrix A:' , A)
print('Matrix B:' , B)
```

## 9.1: Matrix operations

```python
print('A + A = ', A + A)
```

```python
print('A A^T = ', np.dot(A, A.T))
```

```python
print('A^T  A = ', np.dot(A.T, A))
```

```python
print('AB = ', np.dot(A, B))
```

```python
def func(A, B, Lam):
    temp = B - Lam*np.eye(m)
    return np.dot(A, temp)


print('A(B − λI) for λ:12 =', func(A,B, 12))
```

## 9.2: Solving a linear system

```python
b = np.random.rand(m,1)
print('x of Bx=b, where b is a random vector ', np.linalg.solve(B, b))
```

## 9.3: Norms

```python
print('Frobenius norm: ', np.linalg.norm(A))
print('max(sum(abs(x), axis=1)): ', np.linalg.norm(B, ord=np.inf))
print('2-norm (largest sing. value): ', np.linalg.norm(B, ord=2))
print('smallest singular value: ', np.linalg.norm(B, ord=-2))
```

## 9.4: Power iteration

```python
import time

def timer(funcx):
    def wrapper():
        start = time.time()
        funcx()
        end = time.time()
        print('Opertaion takes', end - start, 'seconds')

    return wrapper


@timer
def func4():
    Z = np.random.normal(2,2, (n, n))
    w, v = np.linalg.eig(Z)
    print('Largest eigenvalue: ', w.max())
    print('Corresponding eigenvector: ', v) 

func()

```

## 9.6: Nearest neighbor

```python
def func6(A, z):
    A = np.asarray(A)
    idx = (np.abs(A - z)).argmin()
    return A[idx]

A = np.random.rand(10)
z = 0
print('For matrix A:\n', A, '\nnearest element to z=0 is :', func6(A, z))
```
